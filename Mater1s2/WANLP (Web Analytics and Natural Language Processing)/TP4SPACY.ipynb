```python
import spacy
print(spacy.__version__)
#from spacy.lang.fr import French
#from spacy import displacy
nlp = spacy.load('fr_core_news_sm')
sent= "Le ministre de l'conomie a annonc  Alger des rformes importantes pour stimuler l'investissement dans le secteur des technologies de l'information et de la communication, ce qui a suscit un vif enthousiasme parmi les entrepreneurs algriens. Cependant, certains experts restent sceptiques quant  l'impact de ces mesures sur le march du travail. Par ailleurs, la banque centrale a rvis ses taux d'intrt pour encourager l'pargne, une dcision qui a t accueillie avec prudence par les acteurs financiers. En outre, la ville d'Oran se prpare activement  accueillir le prochain sommet international sur les nergies renouvelables, un vnement qui promet de mettre en lumire les avances de l'Algrie dans ce domaine. Durant la confrence, le ministre a lev son verre en l'honneur de l'innovation, tandis qu'un assistant ajustait le poignet de sa chemise avant de lui remettre un livre contenant la lettre officielle d'invitation."
doc = nlp(sent) 
print(doc)
for t in doc:
    print(t.text,'=>',t.pos_)
    
    
print('\n')



mots =["rformes","enthousiasme","pargne"]
for t in mots:
    token=nlp(t)[0]
    print(token.text, token.lemma_, token.prefix_, token.suffix_)
```
